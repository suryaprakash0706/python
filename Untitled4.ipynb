```python
import pandas as pd
import matplotlib.pyplot as plt


data = [['E1', 'M', 30, 100, 'Normal', 500],
['E2', 'F', 35, 75, 'Overweight', 350],
['E3', 'M', 25, 125, 'Underweight', 750],
['E4', 'F', 40, 50, 'Obesity', 400],
['E5', 'M', 20, 200, 'Fit', 1000]]
df = pd.DataFrame(data, columns = ['Emp_ID', 'Gender', 'Age', 'Sales','BMI', 'Income'])
df.hist()
plt.show()
```

```python
df.plot.bar()
plt.bar(df['Age'], df['Sales'])
plt.xlabel("Age")
plt.ylabel("Sales")
plt.show()
```

```python
df.plot.box()
plt.boxplot(df['Income'])
plt.show()
```

```python
plt.pie(df['Age'], labels = {"A", "B", "C", "D", "E"},
autopct ='% 1.1f %%', shadow = True)
plt.show()
```

```python
plt.scatter(df['Income'], df['Age'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]
data = list(zip(x, y))
linkage_data = linkage(data, method='ward', metric='euclidean')
dendrogram(linkage_data)
plt.show()
```
